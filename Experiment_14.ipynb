```python
import pandas as pd
import numpy as np
data = {
    'Product': ['Laptop','Mobile','Tablet','Headphones','Smartwatch','Camera'],
    'Price': [55000,20000,30000,2000,8000,45000],
    'Unit_sold':[25,60,35,80,50,15]
     }
df=pd.DataFrame(data)
print(df)
```

```python
# Data Binning
bins1 = [0,10000,30000,60000]
labels1 = ['Low','Medium','High']
df['Price_Category'] = pd.cut(df['Price'],bins=bins1,labels=labels1)
print(df)
```

```python
#Binning UNits sold
bins2 = [0,30,60,100]
labels2 = ['Low sales','Medium sales','High sales']
df['Sales_Category'] = pd.cut(df['Unit_sold'],bins=bins2,labels=labels2)
print(df)
```

```python
#Data Formatting
# Convert Data Type
print(df.dtypes)
df['Unit_sold'] = df['Unit_sold'].astype(float)
print(df.dtypes)
```

```python
#convert text to uppercase
df['Product'] = df['Product'].str.upper()
print(df)
```

```python
# rounding off
df['Price'] = df['Price'].round(2)
print(df)
```

```python
#Additional operation
#sort data
sorted_df = df.sort_values(by='Price')
print(sorted_df)
```

```python
#sort data
sorted_df = df.sort_values(by='Price',inplace=True)
df
```

```python
#sorytng data in decending order
sorted_df = df.sort_values(by='Price',ascending=False)
print(sorted_df)
```

```python
#DIsplay unique values
unique_Price_Category = df['Price_Category'].unique()
print(unique_Price_Category )
```

```python
import pandas as pd
import numpy as np
data = {
    'Order_ID':['01','02','03','04','05','06','07','08'],
    'Order_Value':[250,450,120,600,300,150,500,220],
    'Delivery_Time':[30,45,25,60,35,20,50,28],
    'Distance_km':[5,8,3,10,6,2,9,4]
}
df = pd.DataFrame(data)
print(df)
```

```python
#Data binning
bins = [0,200,400,700]
labels = ['low','medium','high']
df['Order_Value_Category'] = pd.cut(df['Order_Value'],bins=bins,labels=labels)
print(df)
```

```python
# binning Delievery time
bins = [0,30,45,75]
label2 = ['Fast','medium','slow']
df['Delivery_speed'] = pd.cut(df['Delivery_Time'],bins=bins,labels=label2)
print(df)
```

```python
#binning distanc_km
bins = [0,4,7,10] # Added an extra bin edge to match the number of labels
label3 = ['short','medium','large']
df['Distance_Category'] = pd.cut(df['Distance_km'],bins=bins,labels=label3)
print(df)
```

```python
# count of categories
print(df['Order_Value_Category'].value_counts())
print(df)
```

```python
#convert into upper case Order_value_category
df['Order_Value_Category'] = df['Order_Value_Category'].str.upper()
print(df)
```

```python
#sort on distance_km
sorted_df = df.sort_values(by='Distance_km')
print(sorted_df)
```
